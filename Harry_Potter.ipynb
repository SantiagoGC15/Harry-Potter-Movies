```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
df=pd.read_csv('/content/Movies.csv')
df
```

```python
Data1=df.sort_values('Budget')
plt.plot(Data1['Release Year'],Data1['Budget'],'k')
for i in range(len(Data1['Release Year'])):
  plt.scatter(Data1['Release Year'][i],Data1['Budget'][i],label='Harry Potter %s' % i)
plt.title('Presupuesto por Película')
plt.xlabel('Año')
plt.ylabel('Presupuesto (US$)')
plt.grid()
plt.legend()
plt.show()
```

```python
Data1=df.sort_values('Box Office')
plt.plot(Data1['Release Year'],Data1['Box Office'],'k')
for i in range(len(Data1['Release Year'])):
  plt.scatter(Data1['Release Year'][i],Data1['Box Office'][i],label='Harry Potter %s' % i)
plt.title('Taquilla por Película')
plt.xlabel('Año')
plt.ylabel('Taquilla (US$)')
plt.grid()
plt.legend()
plt.show()
```

```python
Data1=df.sort_values('Runtime')
plt.plot(Data1['Release Year'],Data1['Runtime'],'k')
for i in range(len(Data1['Release Year'])):
  plt.scatter(Data1['Release Year'][i],Data1['Runtime'][i],label='Harry Potter %s' % i)
plt.title('Tiempo de Ejecución')
plt.xlabel('Año')
plt.ylabel('Tiempo (Min)')
plt.grid()
plt.legend()
plt.show()
```

```python
df2=pd.read_csv('/content/Spells.csv')
df2=df2.dropna(axis='columns')
df2
```

```python
Ef=[]
for n in range(len(df2['Effect'])):
  Ef.append(str(df2['Effect'][n]))

Effect=' '.join([str(elem) for elem in Ef])
```

```python
cloud = WordCloud(max_font_size=50,
                  max_words=100,
                  colormap='Reds',
                  background_color="black").generate_from_text(Effect)
plt.figure(figsize=(10,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```

```python
Spe=[]
for n in range(len(df2['Spell Name'])):
  Spe.append(str(df2['Spell Name'][n]))

Spell=' '.join([str(elem) for elem in Spe])
```

```python
cloud1 = WordCloud(max_font_size=50,
                  max_words=100,
                  colormap='Reds',
                  background_color="black").generate_from_text(Spell)
plt.figure(figsize=(10,8))
plt.imshow(cloud1)
plt.axis('off')
plt.show()
```

```python
Inc=[]
for n in range(len(df2['Incantation'])):
  Inc.append(str(df2['Incantation'][n]))

Incantation=' '.join([str(elem) for elem in Inc])
  
```

```python
#https://towardsdatascience.com/how-to-create-beautiful-word-clouds-in-python-cfcf85141214
cloud = WordCloud(max_font_size=50,
                  max_words=100,
                  colormap='Reds',
                  background_color="black").generate_from_text(Incantation)
plt.figure(figsize=(10,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
